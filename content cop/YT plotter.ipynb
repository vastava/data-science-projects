```python
import pandas as pd
import csv

cc = pd.read_csv('contentcop_v2.csv')
cc
```

```python
names = """Jinx
Fine Bros
Keemstar
HowToPRANKItUp
LeafyIsHere
Tana Mongeau
Ricegum""".splitlines()
names
```

```python
colors = ['#8c564b', '#ff7f0e', '#2ca02c', 'steelblue', '#d62728', '#9467bd', 'lightseagreen']
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(6.4*1.5,3.6*1.5))
for i in range(len(names)):
#     pd.to_numeric(df.Subs).plot.line()
    df = cc[cc.Name==names[i]]
    plt.plot(np.arange(len(df.index)), pd.to_numeric(df.Subs), label = df.Name.values[0], color = colors[i])
    plt.axvline(x=30, color='red', linestyle='dashed',linewidth=0.5)
    plt.ylabel('Change in Subscribers by Day')
    plt.xlabel("Day (0 is 1 month before Content Cop's release)")
    plt.legend()
plt.annotate('Content Cop released', xy=(30, 20000), ha='center', size=10)
plt.title('The Impact of Content Cop')
plt.savefig('cc_label.svg')
#     df.plot.line(x=np.arange(62), y=pd.to_numeric(df.Subs))
```

```python
plt.figure(figsize=(6.4*1.5,3.6*1.5))
for i in range(len(names)):
#     pd.to_numeric(df.Subs).plot.line()
    df = cc[cc.Name==names[i]]
    plt.plot(np.arange(len(df.index)), pd.to_numeric(df.Total).pct_change(), label = df.Name.values[0], color = colors[i])
    plt.axvline(x=30, color='red', linestyle='dashed',linewidth=0.5)
    plt.ylabel('Percent Change in Subscribers by Day')
    plt.xlabel("Day (0 is 1 month before Content Cop's release)")
    plt.legend()
plt.annotate('Content Cop released', xy=(30, 0.015), ha='center', size=10)
plt.title('The Impact of Content Cop')
plt.savefig('pct_cc.svg')
```

```python
plt.figure(figsize=(6.4*1.5,3.6*1.5))
for i in range(len(names)):
#     pd.to_numeric(df.Subs).plot.line()
    df = cc[cc.Name==names[i]]
    plt.plot(np.arange(len(df.index)), pd.to_numeric(df.Total), label = df.Name.values[0], color = colors[i])
    plt.axvline(x=30, color='red', linestyle='dashed',linewidth=0.5)
    plt.ylabel('Change in Total Subscribers')
    plt.xlabel("Day (0 is 1 month before Content Cop's release)")
    plt.legend()
# plt.annotate('Content Cop released', xy=(30, 20000), ha='center', size=10)
plt.title('The Impact of Content Cop')
plt.savefig('tot_cc.svg')
```

```python
plt.figure(figsize=(6.4*1.5,3.6*1.5))
for i in range(len(names)):
#     pd.to_numeric(df.Subs).plot.line()
    df = cc[cc.Name==names[i]]
    val = pd.to_numeric(df.Total).values[0]
    plt.plot(np.arange(len(df.index)), pd.to_numeric(df.Total)/val, label = df.Name.values[0], color = colors[i])
    plt.axvline(x=30, color='red', linestyle='dashed',linewidth=0.5)
    plt.ylabel('Percent Change in Total Subscribers from first day')
    plt.xlabel("Day (0 is 1 month before Content Cop's release)")
    plt.legend()
plt.annotate('Content Cop released', xy=(30, 1.25), ha='center', size=10)
plt.title('The Impact of Content Cop')
plt.savefig('tot_cc_pct.svg')
```
