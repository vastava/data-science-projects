```python
#initialize dataframe
df_temp = pd.DataFrame()

#populate date and average temperature fields (cast string date to datetime and convert temperature from tenths of Celsius to Fahrenheit)
df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates_temp]
df_temp['minTemp'] = [float(v)/10.0*1.8 + 32 for v in temps]
```

```python
df_temp['year'] = pd.DatetimeIndex(df_temp['date']).year
df_temp['month'] = pd.DatetimeIndex(df_temp['date']).month
```

```python
def label_season (row):
    if row['month'] <= 5 and row['month'] >= 3:
        return 'Spring'
    else:
        if row['month'] <= 8 and row['month'] >= 6:
            return 'Summer'
        else:
            if row['month'] <= 11 and row['month'] >= 9:
                return 'Fall'
            else:
                return 'Winter'

df_temp['season'] = df_temp.apply(lambda row: label_season(row), axis=1)
```
